```python
"""绘制线性回归基函数"""

import numpy as np
import matplotlib.pyplot as plt

from linear_regress import gaussian_function, sigmoid_function, polynominal_function
```

```python
x = np.arange(-1, 1, 0.01)
mus = np.arange(-1, 1, 0.2)
sigma = 0.1
degrees = np.arange(1, 10, 1)
```

```python
fig = plt.figure(num=1, figsize=(12, 4))
ax1 = plt.subplot(1, 3, 1)
ax2 = plt.subplot(1, 3, 2)
ax3 = plt.subplot(1, 3, 3)
for mu in mus:
    ax1.plot(x, gaussian_function(x, mu, sigma))
    ax2.plot(x, sigmoid_function(x, mu, sigma))
for degree in degrees:
    ax3.plot(x, polynominal_function(x, degree))

ax1.set_xlim(-1, 1)
ax1.set_ylim(0, 1)
ax1.set_xticks([-1, 0, 1])
ax1.set_yticks([0, 0.5 , 1])
ax1.tick_params(direction='in')

ax2.set_ylim(0, 1)
ax2.set_xticks([-1, 0, 1])
ax2.set_yticks([0, 0.5, 1])
ax2.set_xlim(-1, 1)
ax2.tick_params(direction='in')

ax3.set_xlim(-1, 1)
ax3.set_xticks([-1, 0, 1])
ax3.set_ylim(-1, 1)
ax3.set_yticks([-1, 0, 1])
ax3.tick_params(direction='in')

plt.tight_layout()
plt.show()
```
